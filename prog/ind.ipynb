### Построить график, заданный системой уравнений
$$ \left\{
    \begin{array}\\
    x = cos(u)*u*(1 +cos(\frac{v}{2}));\\
    y = u/2*sin(v);\\
    z = (sin(u)*u)*(1+cos(\frac{v}{2}))
    \end{array}
\right.
$$
$
0 \leq u \leq 2\pi, 0 \leq v \leq 36\pi
$

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
u, v = np.mgrid[0:2*np.pi:20j, 0:36*np.pi:10j]
x = np.cos(u) * u * (1 + np.cos(v / 2))
y = u / 2 * np.sin(v)
z = (np.sin(u) * u) * (1 + np.cos(v / 2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')

plt.show()
```
